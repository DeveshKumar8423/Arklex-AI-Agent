```python
import json

config = {
    "agent_name": "CustomerSupportBot",
    "objectives": ["Answer customer queries", "Retrieve order status"],
    "domain": "E-commerce",
    "resources": {
        "database": "orders_db",
        "api_keys": {
            "openai": "your_api_key_here"
        }
    }
}

with open("config.json", "w") as f:
    json.dump(config, f)

print("Config file created successfully!")
```

```python
from langchain.tools import Tool

def fetch_order_status(order_id):
    return f"Order {order_id} is currently in transit."

order_status_tool = Tool(
    name="OrderStatus",
    func=fetch_order_status,
    description="Fetches order status by order ID"
)

print("Worker module created successfully!")
```
