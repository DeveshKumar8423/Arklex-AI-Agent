```python
!pip install langchain groq

import os
from langchain.chat_models import ChatOpenAI
from langchain.schema import HumanMessage

os.environ["GROQ_API_KEY"] = "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"

llm = ChatOpenAI(
    openai_api_key=os.getenv("GROQ_API_KEY"),
    model_name="llama3-8b-8192",
    openai_api_base="https://api.groq.com/openai/v1"
)

def customer_service_chat():
    print("\n🔹 Customer Service Chatbot (Type 'exit' to stop)\n")
    while True:
        try:
            user_input = input("You: ")
            if user_input.lower() == "exit":
                print("Chatbot: Goodbye! ")
                break
            response = llm.invoke([HumanMessage(content=user_input)])
            print(f"Chatbot: {response.content}\n")
        except Exception as e:
            print(f"Error: {e}")

customer_service_chat()
```
